```python
# 최대 값을 찾는 알고리즘

# 첨자 index      0  1  2  3  4
# 데이터          12 13 11 14 10

arr = [12,13,11,14,10]

max = arr[0]

for i in arr : 
  if i > max:
    max = i

print(max)

```

```python
# 유명한 알고리즘

# 기본적인 처리 절차를 가지고 있는 좋은 알고리즘이다.
# 좋은 프로그램을 만들기 위한 사고방식이나 힌트가 많이 포함되어 있다.
# 유명한 알고리즘을 학습하는 것은 프로그래밍 기술 향상에 커다란 도움이 된다.

# 탐색 Search
# Linear Search 선형 탐색법 - 앞에서부터 순서대로 찾는다.
# Binary Search 이진 탐색법 - 범위를 절반씩 줄여서 찾는다.
# Hash Search 해시 탐색법 - 검색 즉시 찾는다. 단, 저장할 때 가공해서 저장한다.

# 정렬 Sort
# Selection Sort 단순 선택법 - 최소값을 선택하여 앞부터 순서대로 나열한다.
# Bubble Sort 단순 교환법 - 옆에 있는 데이터를 교환하면서 자리를 바꾼다.
# Insertion Sort 단순 삽입법 - 데이터를 올바른 위치에 삽입하면서 자리를 바꾼다.
# Quick Sort 퀵정렬 - 기준 데이터를 기반으로 대소 분할을 반복하여 자리를 바꾼다.

# 에라토스테네스의 체 Sieve of Eratosthenes - 소수 구하는 알고리즘
# 유클리드 알고리즘 - 최대 공약수를 구하는 알고리즘
```

```python
# Linear Search 선형 탐색법

# 보관된 데이터를 맨 앞부터 순서대로 원하는 값을 찾을 때까지 탐색한다.
# 탐색처리는 반복 구조를 사용하여 기술한다.
# 반복 구조에서는 종료 조건을 잊지 말자. 

첨자(index(방번호))     0  1  2  3  4  
데이터                  4  2  3  5  1    찾을 데이터 : 5

```

탐색_선형탐색법.jpg

```python
arr = [4,2,3,5,1]

for i in range(0,5):           # range는 인덱스(방번호), arr는 값이 나옴
  if arr[i] == 5:
    print(f'{i}번째 요소가 일치')
    break
```

```python
# 답 1

arr = [4,2,3,5,1]

for i in range(0,5): 
  if arr[i] == 5:
    print(f'{i}번째 요소가 일치')
    break
  elif i ==4:
    print('찾지 못했습니다.')
```

```python
# 답 2

def linear_search(no, list):
  for i in range(len(list)):
    if no == list[i]:
      return print(f'{i}번째 요소가 일치')
    return print('찾지 못했습니다.')
  

print(linear_search(5, [4,2,3,5,1]))
```

```python
def test():
  print('test()함수의 첫 줄입니다.')
  try:
    print('try 구문이 실행되었습니다.')
    return
    print('try구문의 return 아래입니다.')
  except:
    print('except구문이 실행되었습니다.')
  else:
    print('else구문이 실행되었습니다.')
  finally:
    print('finally구문이 실행되었습니다.')
  print('test()함수의 마지막 줄입니다.')

test()
```
